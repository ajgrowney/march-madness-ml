# Data Cleaning

- If height is not a valid "X'Y" or "X-Y", assign it 72 (6 foot)
- If yr is not valid, assign 'Fr'

```python
def height_to_int(ht, char):
    ft, inches = ht.split(char)
    return int(ft) * 12 + int(inches)

def convert_height(ht):
    try:
        try:
            if ht == "-" or pd.isna(ht) or ht is None or ht == "None":
                return None
            return height_to_int(ht, "-")
        except:
            return height_to_int(ht, "'")
    except Exception as e:
        print(type(ht), ht)
        return 72

def valid_yr(yr):
    if yr in ["Fr", "So", "Jr", "Sr"]:
        return yr
    return "Fr"
```

# Player Style Metrics

StarterPointsPct
BenchPointsPct
BenchMinutesPct
StarterAvgExp
FullTeamExp (players over 10 Min_per)

```python


def play_style_metrics_df(players_df):
    teams_df = pd.DataFrame([], columns = ["StarterPointsPer", "BenchPointsPer", "BenchMinPer", "Height", "StarterExp", "TeamExp"])
    for team, team_players_df in players_df.groupby("team"):
        # Add a boolean column 'starter' based on if the player is in top 5 of Min_per
        team_players_df['starter'] = team_players_df['Min_per'].rank(ascending=False) <= 5
        # Calculate the average points per game for starters and bench players
        starter_pts_percentage = team_players_df[team_players_df['starter']]['pts'].sum() / team_players_df['pts'].sum()
        bench_pts_percentage = team_players_df[~team_players_df['starter']]['pts'].sum() / team_players_df['pts'].sum()
        # Calculate the percentage of team minutes per game from bench players
        bench_min_per = team_players_df[~team_players_df['starter']]['Min_per'].sum() / team_players_df['Min_per'].sum()
        # Calculate the average height of the team (convert "f-in" to inches)
        team_players_df['ht'] = team_players_df['ht'].apply(convert_height)
        team_height = team_players_df['ht'].mean()
        # Calculate the average experience of the starters ( Fr/na: 1, So: 2, Jr: 3, Sr: 4)
        team_players_df['yr'].fillna("Fr", inplace=True)
        starter_exp = team_players_df[team_players_df['starter']]['yr'].map({"Fr": 1, "So": 2, "Jr": 3, "Sr": 4}).mean()
        # Calculate the average experience of the team
        team_exp = team_players_df['yr'].map({"Fr": 1, "So": 2, "Jr": 3, "Sr": 4}).mean()
        teams_df.loc[team] = [starter_pts_percentage, bench_pts_percentage, bench_min_per, team_height, starter_exp, team_exp]
    
    # Create Percentile Ranks for each column
    teams_df["StarterPointsPct"] = teams_df["StarterPointsPer"].rank(pct=True)
    teams_df["BenchPointsPct"] = teams_df["BenchPointsPer"].rank(pct=True)
    teams_df["BenchMinPct"] = teams_df["BenchMinPer"].rank(pct=True)
    teams_df["HeightPct"] = teams_df["Height"].rank(pct=True)
    teams_df["StarterExpPct"] = teams_df["StarterExp"].rank(pct=True)
    teams_df["TeamExpPct"] = teams_df["TeamExp"].rank(pct=True)
    return teams_df
```

```python
TS_DF = pd.read_csv("../data/teamspellings.csv").drop_duplicates(subset="TeamNameSpelling").set_index("TeamNameSpelling")
def get_team_id(style_df):
    style_df["TeamID"] = style_df.index.to_series().str.lower().map(TS_DF["TeamID"])
PLAY_STYLE_DF = pd.DataFrame([], columns = ["Season", "TeamID", "StarterPointsPer", "BenchPointsPer", "BenchMinPer", "Height", "StarterExp", "TeamExp", "StarterPointsPct", "BenchPointsPct", "BenchMinPct", "HeightPct", "StarterExpPct", "TeamExpPct"])
for season in range(2008, 2025):
    print(f"Processing {season}")
    players_season = pd.read_csv(f"{DATA_ROOT}/{season}.csv", header=None, names=headers)
    play_style_season = play_style_metrics_df(players_season)
    play_style_season["Season"] = season
    get_team_id(play_style_season)
    PLAY_STYLE_DF = pd.concat([PLAY_STYLE_DF, play_style_season])

print(PLAY_STYLE_DF)
```
